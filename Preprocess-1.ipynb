# Preprocess the data

In this notebook we fetch the data from all the files in the dataset and create a csv file called new_file.csv to facilitate easy access to data.

```python
import pandas as pd
import numpy as np
import os
```

# This is the path to the dataset

```python
DATA_PATH = "/home/tushar/Downloads/enron_mail_dataset/maildir"
```

# Fetch all the files

```python
listOfFiles = list()
for (dirpath, dirnames, filenames) in os.walk(DATA_PATH):
    listOfFiles += [os.path.join(dirpath, file) for file in filenames]
```

# Fetch all the data

```python
data = []
```

```python
import codecs
```

```python
for i in range(len(listOfFiles)):
    print(i)
    row = []
    row.append(listOfFiles[i])
    with codecs.open(listOfFiles[i], 'r', encoding='utf-8',errors='ignore') as fdata:        
        row.append(fdata.read())    
    data.append(row)
```

# Write the data to the csv

```python
import csv

with open("new_file.csv","w+") as my_csv:
    csvWriter = csv.writer(my_csv,delimiter=',')    
    csvWriter.writerows(data)
```
