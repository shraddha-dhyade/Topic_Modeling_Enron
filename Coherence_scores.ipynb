## Coherence score comparison plot for LSA, LDA and NMF

```
import matplotlib.pyplot as plt
import seaborn as sns

k_values = ['LSA', 'LDA', 'NMF']
coherences = [0.4883421540329757, 0.6454252378603928, 0.7318232187890401]

plt.figure(figsize=(16, 10))

ax = sns.barplot(x=k_values , y=coherences, palette="Greens")
ax.axes.set_title("Coherence score by model",fontsize=30)
ax.set_xlabel("Models",fontsize=30)
ax.set_ylabel("Coherence",fontsize=30)

sns.set(style="whitegrid")
ax.tick_params(axis = 'x', labelsize=25, labelrotation = 90)
ax.tick_params(axis = 'y', labelsize=25)
plt.savefig('/content/drive/My Drive/ALDA-Submission/Plots/Coherence_Scores.png')
plt.show()
```
