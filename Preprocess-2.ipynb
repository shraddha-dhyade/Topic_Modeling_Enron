# Preprocess and preliminary cleaning

In this notebook, we seperate the email data according to different fields in the email and then create a csv named processed_data.csv

```
import pandas as pd
import email
```

# Fetch the new_file.csv

```
dataset = pd.read_csv('/content/drive/My Drive/ALDA-Project/new_file.csv', names = ['filename','content'], header = None)
```

# Fetch the email content

```
messages = list(map(email.message_from_string, dataset['content']))
```

```
print(messages[0])
```

# Fetch the email headers

```
headers = messages[0].keys()
print(headers)
for header in headers:
    dataset[header] = [content[header] for content in messages]
```

```
dataset.head()
```

```
message_contents=[]
for message in messages:
    message_content = ''
    for part in message.walk():
        if part.get_content_type() == 'text/plain':
            message_content = message_content + part.get_payload()
    message_contents.append(message_content)
dataset['message_content'] = message_contents
```

```
dataset.head()
```

```
def split_email_ids(email_ids):
    if email_ids:
        return list(set(email_id.strip(' ') for email_id in email_ids.split(','))) 
    return
```

```
dataset['To'] = dataset['To'].map(split_email_ids)
```

```
dataset['From'] = dataset['From'].map(split_email_ids)
```

```
dataset['user'] = dataset['filename'].map(lambda x: x.split('/')[6])
```

```
dataset.head()
```

```
dataset = dataset.drop(['content', 'Mime-Version', 'Content-Type', 'Content-Transfer-Encoding'], axis=1)
```
